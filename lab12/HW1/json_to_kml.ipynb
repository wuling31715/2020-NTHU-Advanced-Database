```python
df = pd.read_json('data/pm25.json')
```

```python
def get_site(site):
    site_list = site.split('(')
    return site_list[0]
```

```python
pm25_list = list()
pollution = list()
for feeds in df['feeds']:
    site = get_site(feeds['site'])
    gps_lat = feeds['gps_lat']
    gps_lon = feeds['gps_lon']
    pm25 = feeds['pm25']
    pollution = feeds['pollution']
    name = '{} PM2.5: {}'.format(site, pm25)
    pm25_list.append([name, gps_lon, gps_lat])
```

```python
pm25_list
```

```python
import simplekml
kml = simplekml.Kml()
for i in pm25_list:
    x = i[1]
    y = i[2]
    if type(x) == float and type(y) == float:
        kml.newpoint(name=i[0], coords=[(x, y)])
kml.save("data/pm25.kml")
```
